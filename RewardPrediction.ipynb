Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv("trainingData.csv")
print(df)
```

```python
%matplotlib inline
plt.xlabel("timeStamp (seconds)")
plt.ylabel("blockRewards (ethers)")
plt.scatter(df.timestamp,df.blockReward,color="red",marker="*")
```

```python
model = RandomForestRegressor(n_estimators=100,random_state=42)
model.fit(df.timestamp.values.reshape(-1,1),df.blockReward.values.ravel())
```

```python
new_timeStamp = np.array([[1686644543]])
predictedBlockReward = model.predict(new_timeStamp)

plt.xlabel("timeStamp (seconds)")
plt.ylabel("blockRewards (ethers)")
plt.scatter(df.timestamp,df.blockReward,color="red",marker="*",label="Data Points")
plt.scatter(new_timeStamp,predictedBlockReward[0],color="blue",marker="*",label="Predicted Future Block Reward")

print("The predicted Block Reward is : ",predictedBlockReward[0])
```
